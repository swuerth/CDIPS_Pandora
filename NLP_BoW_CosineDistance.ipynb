```python
from scipy import spatial

#use this one if you're inputting sparse arrays that need to be converted
#def cos_sim(your_song_bow_vec, other_song_bow_vec):
#    your_song_bow_vec = your_song_bow_vec.toarray()
#    other_song_bow_vec = other_song_bow_vec.toarray()
#    result = 1 - spatial.distance.cosine(your_song_bow_vec, other_song_bow_vec)
#    return result

#use this one if your input vectors are already converted to arrays (from sparse)
def cos_dis(your_song_bow_vec, other_song_bow_vec):
    result = spatial.distance.cosine(your_song_bow_vec, other_song_bow_vec)
    return result
```

```python
# test function
song1 = 'Bohemian Rhapsody'
song2 = 'Lose Yourself'
bow1 = lyrics_df_with_bow[lyrics_df_with_bow['track_name']==song1].BoW_array.values[0].toarray()
bow2 = lyrics_df_with_bow[lyrics_df_with_bow['track_name']==song2].BoW_array.values[0].toarray()
print 'size bag of words vectors = ', np.shape(bow1), ' and ', np.shape(bow2)
cos_distance = cos_dis(bow1,bow2)
print 'cosine distance between ', song1, ' and ', song2 , ' = '
print cos_distance
```
